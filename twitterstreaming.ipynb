```python
rounds = [['round'+str(x)+'.csv' for x in range(1,13)],
          'prefight.csv',
          'mayweathertweets.csv',
          'pacquiaotweets.csv',
          'fightnighttweets.csv']

fightnight = pd.DataFrame()

for roundx in rounds:
    reader = pd.read_csv(csv, low_memory=False)
    df = reader.fillna('')
    df['Tweet'] = df['Tweet'].map(str) + ',' + df['tweet_2'].map(str) + ',' + df['tweet_3'].map(str) + ',' + df['tweet_4'].map(str) + ',' + df['tweet_5'].map(str) + ',' + df['tweet_6'].map(str) + ',' + ',' + df['tweet_7'].map(str) + ',' + df['tweet_8'].map(str) + ',' + df['tweet_9'].map(str) + ',' + df['tweet_10'].map(str) + ',' + df['tweet_11'].map(str) + ',' + df['tweet_12'].map(str)  + ',' + df['tweet_13'].map(str) 
    df = df.drop([x for x in df.columns[3:-1]], axis=1)
    df = df[['Timestamp','User','Tweet']]
    df = df[[x[-5:] == ' 2015' for x in df['Timestamp']]]
    df['Hashtags'] = [re.findall(r"#(\w+)", tweet) for tweet in df.Tweet]
    df['Hashtags'] = [', '.join(x) for x in df['Hashtags']]
    fightnight = fightnight.append(df, ignore_index=True)
```

```python
all_rounds = pd.DataFrame()

for r in rounds2:
    reader = pd.read_csv(r, low_memory=False)
    df = reader.fillna('')
    df['Tweet'] = df['Tweet'].map(str) + ',' + df['tweet_2'].map(str) + ',' + df['tweet_3'].map(str) + ',' + df['tweet_4'].map(str) + ',' + df['tweet_5'].map(str) + ',' + df['tweet_6'].map(str) + ',' + ',' + df['tweet_7'].map(str) + ',' + df['tweet_8'].map(str) + ',' + df['tweet_9'].map(str) + ',' + df['tweet_10'].map(str) + ',' + df['tweet_11'].map(str) + ',' + df['tweet_12'].map(str)  + ',' + df['tweet_13'].map(str) 
    df = df[['Timestamp','User','Tweet']]
    df = df[[x[-5:] == ' 2015' for x in df['Timestamp']]]
    df['Round'] = ''.join(c for c in r if c.isdigit())
    all_rounds = all_rounds.append(df)
    
all_rounds = all_rounds.reset_index(drop=True)

#handling hashtags
all_rounds['Hashtags'] = [re.findall(r"#(\w+)", tweet) for tweet in all_rounds.Tweet]
#all_rounds['Hashtags'] = [', '.join(x) for x in all_rounds['Hashtags']] #returns list elements as strings

#handling rounds type conversion
all_rounds['Round'] = [int(x) for x in all_rounds['Round']]

#separating all hastags within each tweet
s = all_rounds.apply(lambda x: pd.Series(x['Hashtags']),axis=1).stack().reset_index(level=1, drop=True)
s.name = 'Hashtag'
all_rounds = all_rounds.drop('Hashtags', axis=1).join(s)
all_rounds = all_rounds.fillna('')

#cleaning hashtag spacing
all_rounds['Hashtag'] = [re.sub(r"(\w)([A-Z])", r"\1 \2", x) for x in all_rounds['Hashtag']]
all_rounds['Hashtag'] = ["".join(x).replace(" ", "") if x.isupper() == True else x for x in all_rounds['Hashtag']]
```

```python
all_rounds
```

```python
#isolate all unique hashtags
hashtags = [x for x in all_rounds['Hashtag'] if len(x) > 0]

#add space before all capitalzed letters, e.g. '#ComeOn' >>> 'ComeOn' >>> 'Come On'
hashtags = [re.sub(r"(\w)([A-Z])", r"\1 \2", x) for x in hashtags]

#remove spaces between only upper case hashtags, e.g. '#MGM >>> 'M G M' >>> 'MGM'
hashtags = ["".join(x).replace(" ", "") if x.isupper() == True else x for x in hashtags]
```

```python
tags = pd.DataFrame(hashtags).rename(columns={0:'Hashtags'})
tags['Count'] = 1
tags = tags.groupby('Hashtags').count().sort('Count', ascending=False)
```
